#MDS & heatmap

```r
filtered.counts <- filtered.counts[ , -1]
head(filtered.counts)
```

```r
tissue.col <- c("#000000", "#000000", "#000000", "#000000", "#000000", "#00FF00", "#00FF00", "#00FF00", "#00FF00", "#00FF00")
```

```r
tissue.col
```

```r

correlation = cor(filtered.counts, method = "spearman")
```

```r
heatmap(correlation,
symm = TRUE,
distfun = function(x){as.dist(1-x)},
ColSideColors=tissue.col)
```

```r
age.size <- c(1.4, 1.5, 1.7, 2.0, 3.4, 1.4, 1.5, 1.7, 2.0, 3.4)
```

```r
mds=cmdscale(1-correlation,k=2)
plot(mds[,1],mds[,2],
col=tissue.col,
pch=19,
cex=age.size)
```
